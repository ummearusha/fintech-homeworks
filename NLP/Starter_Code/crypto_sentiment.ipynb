# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.corpus import reuters

from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re

import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Expand the default stopwords list if necessary
print(stopwords.words('english')) 
```

```python
  
# Complete the tokenizer function
#  variable used inside the function not from outside/out of scope.
def tokenizer(text):
    """Tokenizes text."""
    

    # Remove the punctuation
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    words_punc_cleaned = word_tokenize(re_clean)

    # Create a list of the words
    words_lowercased = []
    
    # Convert the words to lowercase
    for word in words_punc_cleaned:
         words_lowercased.append(word.lower())
    
    # Remove the stop words
    sw = set(stopwords.words('english'))
    sw_addon = {'rather', 'blockfi', 'applying'}
    words_stop_removed = [word for word in words_lowercased if word not in sw.union(sw_addon)]
    
    # Lemmatize Words into root words
    lemmatizer = WordNetLemmatizer() 
    tokens = [lemmatizer.lemmatize(word) for word in words_stop_removed]

    return tokens
  
    

```

```python
bitcoin_tokens = []
for bitcoin_article_text in bitcoin_df['text']:
    bitcoin_tokens.append(tokenizer(bitcoin_article_text))
```

```python
print(bitcoin_tokens)
```

```python
# Create a new tokens column for bitcoin
bitcoin_df['tokens'] = bitcoin_tokens
bitcoin_df.head()
```

```python
eth_tokens = []

for ethereum_article_text in ethereum_df['text']:
    eth_tokens.append(tokenizer(ethereum_article_text))
```

```python
print(eth_tokens)
```

```python
# Create a new tokens column for ethereum
ethereum_df['tokens'] = eth_tokens
ethereum_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
def flatten_list_of_lists(lists_of_list):
    flattened_lists = []
    for list in lists_of_list:
        for item in list:
            flattened_lists.append(item)
    return flattened_lists

# flatten all tokens in one list
bitcoin_tokens_flattened = flatten_list_of_lists(bitcoin_tokens)
# Generate the Bitcoin N-grams where N=2
bitcoin_ngrams = ngrams(bitcoin_tokens_flattened, n=2)
```

```python
# flatten all tokens in one list
eth_tokens_flattened = flatten_list_of_lists(eth_tokens)
# Generate the Ethereum N-grams where N=2
ethereum_ngrams = ngrams(eth_tokens_flattened, n=2)
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(10)
```

```python
# Get the top 10 words for Bitcoin
bitcoin_top_10 = token_count(bitcoin_tokens_flattened, N=10)
bitcoin_top_10
```

```python
# Get the top 10 words for Ethereum

eth_top_10 = token_count(eth_tokens_flattened, N=10)
eth_top_10
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
bitcoin_article_texts = bitcoin_df["text"].str.cat()
 
```

```python
# Run the NER processor on all of the text with a header
doc = nlp(u"Bitcoin NER\n" + bitcoin_article_texts)
```

```python
# Render the visualization
displacy.render(doc, style='ent')
```

```python
# List all Entities

for ent in doc.ents: 
    if ent.label_ == 'GPE':
        print(ent.text)
```
